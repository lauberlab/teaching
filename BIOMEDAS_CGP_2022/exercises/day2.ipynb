# Alignment

We aim at the following:

 * extract our novel flaviviral protein sequence
 * extract the HMM profile of the best hit
 * create alignment of new and reference flaviruses based on the pHMM

```python
# extract discovered flavivirus protein sequence
# on Day 1 we found that the sequence with ID '2049' is from a flavivirus
!grep -A 1 '>2049' ../Day1/SRR5945842_plass_clustered_rep_seq.fasta > 2049.fasta

# extract Flavi pHMM from Pfam database
# on Day 1 we found that this is the profile with ID 'Flavi_DEAD'
!hmmfetch ../Day1/Pfam-A.hmm Flavi_DEAD > Flavi_DEAD.hmm
```

```python
# add sequence to Pfam alignment
#  requires downloading of Flavi_DEAD seed alignment in Fasta format saved in file Flavi_DEAD.fasta
!hmmalign --mapali Flavi_DEAD.fasta --outformat A2M Flavi_DEAD.hmm 2049.fasta > Flavi_DEAD_2049.fasta
```
